```python
import pandas as pd
```

```python
df = pd.read_csv('data/hanoi-air-quality.csv')
df['date'] = pd.to_datetime(df['date'], errors='coerce', format='%Y/%m/%d')
df.set_index('date', inplace=True)
df.sort_index(inplace=True)
df
```

```python
# strip whitespace from column names (if any)
df.columns = df.columns.str.strip()

# treat empty/whitespace strings as missing, then keep only rows without any missing values
df_clean = df.replace(r'^\s*$', pd.NA, regex=True)
df_no_null = df_clean.dropna(how='any').reset_index(drop=False)

df_no_null
```
