```python
import pandas as pd
```

```python
df = pd.read_csv('data\weatherbit_final.csv')
df
```

```python
df.info()
```

```python
df.drop(columns=['pm10', 'o3', 'so2', 'no2', 'co'], inplace=True)
df.dropna(inplace=True)
```

```python
df.info()
```

```python
df['lon'].unique()
```

```python
# check for duplicate timestamps
dupe_mask = df['timestamp'].duplicated(keep=False)
has_dupes = dupe_mask.any()

print("Any duplicate timestamps:", has_dupes)

if has_dupes:
    # show counts for timestamps that appear more than once
    dup_counts = df['timestamp'].value_counts()
    dup_timestamps = dup_counts[dup_counts > 1]
    print("\nTimestamps with more than one row (timestamp: count):")
    print(dup_timestamps)
    # display all rows that have duplicated timestamps
    duplicated_rows = df[dupe_mask].sort_values('timestamp')
    print("\nExample duplicated rows:")
    print(duplicated_rows.head(20))
```

```python
# find counts per (lat, lon)
counts = df.groupby(['lat', 'lon']).size().rename('count').sort_values(ascending=False)

# most frequent pair(s)
max_count = counts.max()
most_frequent = counts[counts == max_count]

print("Most frequent (lat, lon) pair(s) and count:")
print(most_frequent)

print("\nTop 5 (lat, lon) pairs by count:")
print(counts.head(5))
```
